# 多元1次

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler

# 读取Excel数据
data = pd.read_excel('no_error.xlsx')

# 定义自变量X和因变量y
X = data[['母亲年龄','教育程度','CBTS', 'EPDS', 'HADS']]
#X = data[['母亲年龄','CBTS', 'EPDS', 'HADS']]
y = data['婴儿行为特征']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# 添加截距项
#X = sm.add_constant(X_scaled)
poly = PolynomialFeatures(degree=1, interaction_only=False, include_bias=False)
X_poly = poly.fit_transform(X_scaled)
# 建立多元回归模型
model = sm.OLS(y, X_poly).fit()

# 输出模型摘要
print(model.summary())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 读取数据
data = pd.read_excel('no_error.xlsx')

# 处理缺失值（如果有）
data = data.dropna()

# 分割自变量和因变量
X = data[['母亲年龄', '婚姻状况', '教育程度', '妊娠时间', '分娩方式', 'CBTS', 'EPDS', 'HADS']]
y = data['婴儿行为特征']

# 标准化自变量
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 使用PCA降维为2个主成分
n_components = 8
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)

# 建立2元2次交叉回归模型
poly = PolynomialFeatures(degree=4, interaction_only=False, include_bias=False)
X_poly = poly.fit_transform(X_scaled)

# 创建回归模型
model = LinearRegression()

# 拟合回归模型
model.fit(X_poly, y)

# 预测
y_pred = model.predict(X_poly)

# 计算R方
r2 = r2_score(y, y_pred)
print("R方:", r2)
coefficients = model.coef_
intercept = model.intercept_

print("系数：")
for i, coef in enumerate(coefficients):
    print(f"β{i}：{coef}")
```

# 灵敏性分析

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# 读取Excel数据
data = pd.read_excel('no_error.xlsx')

# 定义自变量X和因变量y
X = data[['母亲年龄', '教育程度', 'CBTS', 'EPDS', 'HADS']]
y = data['婴儿行为特征']

# 定义扰动函数，添加扰动项
def perturb_data(X, percentage):
    perturbation = X * percentage / 100.0
    return X + np.random.uniform(-perturbation, perturbation, X.shape)

# 标准化处理
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 添加多项式特征
poly = PolynomialFeatures(degree=3, interaction_only=False, include_bias=False)
X_poly = poly.fit_transform(X_scaled)

# 建立多元回归模型
model = sm.OLS(y, X_poly).fit()

# 输出模型摘要
print(model.summary())

# 进行敏感性分析并作图
perturbation_levels = [5, 10]
misclassification_rates = []
for perturb_level in perturbation_levels:
    perturbed_X = perturb_data(X_scaled, perturb_level)
    perturbed_X_poly = poly.transform(perturbed_X)
    perturbed_y_pred = model.predict(perturbed_X_poly)
    
    # 选择误差小于0.5的预测点
    mask = np.abs(perturbed_y_pred - y) < 0.57
    perturbed_y_pred_filtered = perturbed_y_pred[mask]
    y_filtered = y[mask]
    
    # 计算混淆矩阵并获取错误分类的百分比
    cm = confusion_matrix(y_filtered, np.round(perturbed_y_pred_filtered))
    misclassification_rate = 1 - np.sum(np.diag(cm)) / np.sum(cm)
    misclassification_rates.append(misclassification_rate)

# 绘制敏感性分析图
plt.figure()
plt.plot(perturbation_levels, misclassification_rates, marker='o')
plt.xlabel('Perturbation Percentage')
plt.ylabel('Misclassification Rate')
plt.title('Sensitivity Analysis for Misclassification Rate')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# 读取Excel数据
data = pd.read_excel('no_error.xlsx')

# 定义自变量X和因变量y
X = data[['母亲年龄', '教育程度', 'CBTS', 'EPDS', 'HADS']]
y = data['婴儿行为特征']

# 定义扰动函数，添加扰动项
def perturb_data(X, percentage):
    perturbation = X * percentage / 100.0
    return X + np.random.uniform(-perturbation, perturbation, X.shape)

# 标准化处理
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 添加多项式特征
poly = PolynomialFeatures(degree=3, interaction_only=False, include_bias=False)
X_poly = poly.fit_transform(X_scaled)

# 建立多元回归模型
model = sm.OLS(y, X_poly).fit()

# 输出模型摘要
print(model.summary())

# 进行敏感性分析并作图
perturbation_levels = [5, 10]
error_threshold = 0.25
actual_y = []
predicted_y = []

for perturb_level in perturbation_levels:
    perturbed_X = perturb_data(X_scaled, perturb_level)
    perturbed_X_poly = poly.transform(perturbed_X)
    perturbed_y_pred = model.predict(perturbed_X_poly)
    
    # 记录误差小于阈值的点
    errors = np.abs(perturbed_y_pred - y)
    mask = errors < error_threshold
    actual_y.extend(y[mask])
    predicted_y.extend(perturbed_y_pred[mask])

# 绘制分类任务的敏感性分析图
plt.figure()
plt.scatter(actual_y, predicted_y, alpha=0.5)
plt.xlabel('Actual y')
plt.ylabel('Predicted y')
plt.title('Sensitivity Analysis for Classification')
plt.grid(True)
plt.show()
```
